```python
import glob                   #Libreria que devuelve una lista con las entradas que coincidan con el patrón especificado en pathname.
import pandas as pd           #Libreira que permiyte
import xml.etree.ElementTree as ET
from datetime import datetime

#definiendo funciones de extraccion de datos y probandolas

def extraer_csv(archivo_csv):
    archivo_dataframe=pd.read_csv(archivo_csv)
    return archivo_dataframe

def extraer_json(archivo_json):
    archivo_dataframe=pd.read_json(archivo_json,lines=True)
    return archivo_dataframe

def extraer_xml(archivo_xml):
    archivo_dataframe = pd.DataFrame(columns=["name", "height", "weight"])
    arbol = ET.parse(archivo_xml)
    raiz = arbol.getroot()
    for person in raiz:
        name = person.find("name").text
        height = float(person.find("height").text)
        weight = float(person.find("weight").text)
        archivo_dataframe = archivo_dataframe.append({"name":name, "height":height, "weight":weight}, ignore_index=True)
    return archivo_dataframe


def extraer():
    datos_extraidos=pd.DataFrame(columns=['name','height','weight'])
    for archivocsv in glob.glob('*.csv'):
        datos_extraidos=datos_extraidos.append(extraer_csv(archivocsv),ignore_index=True)
        
    for archivojson in glob.glob('*.json'):
        datos_extraidos=datos_extraidos.append(extraer_json(archivojson),ignore_index=True)
    
    for archivoxml in glob.glob('*.xml'):
        datos_extraidos=datos_extraidos.append(extraer_xml(archivoxml),ignore_index=True)
        
    return datos_extraidos

extraer()





```
