```python
import glob                   #Libreria que devuelve una lista con las entradas que coincidan con el patr√≥n especificado en pathname.
import pandas as pd           #Libreira que permiyte
import xml.etree.ElementTree as ET
from datetime import datetime

tmpfile    = "temp.tmp"               # file used to store all extracted data
archivocargar   = "archivo_a_cargar.txt"            # all event logs will be stored in this file
archivo_destino = "datos_transformados.csv"   # file where transformed data is stored

#definiendo funciones de extraccion de datos y probandolas

def extraer_csv(archivo_csv):
    archivo_dataframe=pd.read_csv(archivo_csv)
    return archivo_dataframe

def extraer_json(archivo_json):
    archivo_dataframe=pd.read_json(archivo_json,lines=True)
    return archivo_dataframe

def extraer_xml(archivo_xml):
    archivo_dataframe = pd.DataFrame(columns=["car_model", "year_of_manufacture", "price","fuel"])
    arbol = ET.parse(archivo_xml)
    raiz = arbol.getroot()
    for row in raiz:
        car_model = row.find("car_model").text
        year_of_manufacture = row.find("year_of_manufacture").text
        price = float(row.find("price").text)
        fuel = row.find("fuel").text
        archivo_dataframe = archivo_dataframe.append({"car_model":car_model, "year_of_manufacture":year_of_manufacture, "price":price, "fuel":fuel}, ignore_index=True)
    return archivo_dataframe


def extraer():
    datos_extraidos=pd.DataFrame(columns=['car_model','year_of_manufacture','price','fuel'])
    for archivocsv in glob.glob('*.csv'):
        datos_extraidos=datos_extraidos.append(extraer_csv(archivocsv),ignore_index=True)
        
    for archivojson in glob.glob('*.json'):
        datos_extraidos=datos_extraidos.append(extraer_json(archivojson),ignore_index=True)
    
    for archivoxml in glob.glob('*.xml'):
        datos_extraidos=datos_extraidos.append(extraer_xml(archivoxml),ignore_index=True)
        
    return datos_extraidos

def transformar_datos(row):
    row['price']=round(row.price,2)
    return row


def cargar(archivo_destino,datos_a_cargar):
    datos_a_cargar.to_csv(archivo_destino)
    
def log(message):
    timestamp_format = '%Y-%h-%d-%H:%M:%S' # Year-Monthname-Day-Hour-Minute-Second
    now = datetime.now() # get current timestamp
    timestamp = now.strftime(timestamp_format)
    with open("archivo_a_cargar","a") as f:
        f.write(timestamp + ',' + message + '\n')

log("trabajo ETL comienza")

log("extraccion de datos inicia")
datos_extraidos = extraer()
log("Extract phase Ended")
datos_extraidos

log("iniciando transformacion de datos")
transformar = transformar_datos(datos_extraidos)
log("finalizando transformacion de datos")
transformar

log("iniciando carga de datos")
cargar(archivo_destino,transformar)
log("finalizando carga de datos")

log("trabajo ETL finalizado")
```
