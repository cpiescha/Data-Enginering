```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm
df = pd.read_csv("https://raw.githubusercontent.com/DrueStaples/Population_Growth/master/countries.csv")
spain=df[df['country']=='Spain']
argentina=df[df['country']=='Argentina']
print(spain.head())
print(argentina.head())
anios=spain["year"].unique()
print(anios)
pop_ar=spain["population"].values
pop_sp=argentina["population"].values
graf=pd.DataFrame({"Argentina":pop_ar,"spain":pop_sp},index=anios)
print(graf)
graf.plot(kind="bar")



```
